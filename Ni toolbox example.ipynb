# Generating random data


```python
conf = ni.model.net_sim.SimulationConfiguration()
conf.Nneur = 10
conf.Ntrials = 100
r = ni.model.net_sim.simulate(conf)
data = r.data
```

```python
spikes = where(data.trial(0).data)
plot(spikes[1],spikes[0],'|')
ylim(-0.5,9.5)
```

```python
for d in data.firing_rate():
    ni.tools.plot.plotGaussed(d,20)
```

# Fitting a Model on the Data
The next part of the code creates a model and uses half of the generated data to fit it. The output is saved to an html file and displayed.

```python
model = ni.model.ip.Model({'autohistory_2d': True})
fm = model.fit(data.trial(range(50)))
fm.html_view().render("test_model.html")
from IPython.display import HTML, FileLink, FileLinks
HTML('<iframe src="files/test_model.html" width="100%" height="1000px"></iframe>')
```
